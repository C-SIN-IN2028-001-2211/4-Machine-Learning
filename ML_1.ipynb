## Explorando los datos

Lo primero que hay que hacer en cualquier proyecto de ML es explorar los datos para determinar cuales serán los factores predictivos y cual será el dato a predecir

```python
import pandas as pd

# save filepath to variable for easier access
melbourne_file_path = '/content/drive/MyDrive/Colab Notebooks/datasets/melb_data.csv'

# read the data and store data in DataFrame titled melbourne_data
melbourne_data = pd.read_csv(melbourne_file_path) 

# print a summary of the data in Melbourne data
melbourne_data.describe()
```

```python
melbourne_data.head()
```

```python
melbourne_data.columns
```

Es importante preprocesar el dataset para que tenga los datos adecuados para construir mi modelo de ML. Uno de los problemas a resolver es el caso cuando faltan datos. La solución más simple es eliminar el renglón (o columna) correspondiente. Hay otras soluciones.

```python
melbourne_data = melbourne_data.dropna(axis=0)
```

```python
melbourne_data.describe()
```
